```python
from nltk import tokenize
from nltk import tag
import re

text = "Hello, World! 123 @# 한글_테스트."

# 영문, 숫자, 한글, 공백을 제외한 나머지 특수문자 제거
cleaned_text = re.sub(r'[^a-zA-Z0-9가-힣 ]', '', text)
dic_voca = {}
for row in df['text']:
    # print(index,row)
    text = re.sub(r'[^a-zA-Z0-9 ]', '', row.lower())
    # print(text)
    word_tokens = tokenize.word_tokenize(text)
    sent_tokens = tokenize.sent_tokenize(text)
    pos = tag.pos_tag(word_tokens)

    for token in word_tokens:
        if token in dic_voca:
            # 이미 있는 단어면 카운트 증가
            dic_voca[token] += 1 
        else:
            # 처음 보는 단어면 1로 초기화
            dic_voca[token] = 1
        
# dict(sorted(dic_voca.items(), key=lambda x: x[1], reverse=True))
print(dict(sorted(dic_voca.items(), key=lambda x: x[1], reverse=True)))

    # print(word_tokens)
    # print(sent_tokens)
    # print(pos)
```
